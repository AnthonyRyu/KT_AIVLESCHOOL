## 3.이변량 분석

* 자료의 종류에 맞게 X --> Y 에 대해서 그래프(시각화)와 가설검정(수치화)를 수행하고 결과를 평가합니다.

* 가설검정시 다음의 항목을 참조하여 수행합니다.
    * 유의수준 : 5%
    * 숫자 --> 숫자 : 상관분석
    * 범주 --> 범주 : 카이제곱검정
    * 범주 --> 숫자 : t검정, 분산분석
    * 숫자 --> 범주 : 로지스틱 회귀모형을 통해, 회귀계수의 P.value로 검정을 수행합니다.
    
상관계수 (우리가 배운 기준)
- 0.5 ~ 1 : 강한
- 0.2 ~ 0.5 : 중간
- 0.1 ~ 0.2 : 약간
- 0 ~ 0.1 : 없음
- 음수 똑같음

p-value 기준을 0.05로 가져죠? 5%
p-value가 0.05보다 크다 == 귀무가설을 기각 하지 못한다 == 대립가설이 틀렸다(우리가 원하는 바는 대립가설)

- 귀무가설을 기각하려면 우리가 싫어하는 귀무가설을 입증할 증거가 나올 확률이 5% 이하여야 함
- p-value가 그 귀무가설을 입증할 증거가 나올 확률을 의미함


```python
x = data["총 인구수"]
y = data["총 이동인구"]

# 피어슨 상관 계수 계산
pearsonr = stats.pearsonr(x, y)

# 상관 계수 출력
print("피어슨 상관 계수:", pearsonr[0])

# 유의도 출력
print("p-value:", pearsonr[1])

# p-value가 0.05보다 작으므로 총 인구수와 총 이동인구는 상관관계가 있습니다.
```

```python
print(data["총 인구수"].corr(data["총 이동인구"]))

# 양의 상관관계를 나타냅니다. 즉, "총 인구수"가 증가할수록 "총 이동인구"도 증가하는 경향이 있다는 것을 의미합니다.
# 상관계수의 절댓값 0.55로 강한 상관관계를 나타냅니다.
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(data[["총 인구수"]], data["총 이동인구"])
print(model.coef_, model.intercept_)


# 계수 (37.90495228): "총 인구수"가 1명 증가할 때 "총 이동인구"는 평균적으로 37.90495228명 증가한다는 것을 의미합니다.

# 절편 (6395242.107492372): "총 인구수"가 0명일 때 "총 이동인구"는 평균적으로 6395242.107492372명이라는 것을 의미합니다.
# 하지만 실제로 "총 인구수"가 0명일 수는 없으므로 이 값은 실제 의미를 가지지 않습니다.

# 총 인구수많으면 총 이동인구(유동인구)도 증가한다.
# 양의 상관관계를 나타냅니다. 즉, "총 인구수"가 증가할수록 "총 이동인구"도 증가하는 경향이 있다는 것을 의미합니다.
```
